```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 2
uniform_samples_01 = np.random.uniform(0,1,1000)
normal_distribution_01 = np.random.normal(0,1,1000)
normal_distribution_010 = np.random.normal(0,10,1000)
beta_distribution_11 = np.random.beta(1,1,1000)
beta_distribution_22 = np.random.beta(2,2,1000)
beta_distribution_39 = np.random.beta(3,9,1000)
two_dimensional_1001 = np.random.multivariate_normal([0,0], [[1,0],[0,1]], 5000)
two_dimensional_2112 = np.random.multivariate_normal([0,0], [[2,1],[1,2]], 5000)
```

```python
# 3
bins = 30
density = True
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=[12,6])
ax[0].hist(uniform_samples_01, bins=bins, density=density)
ax[1].hist(normal_distribution_01, bins=bins, density=density)
ax[2].hist(normal_distribution_010, bins=bins, density=density)
plt.tight_layout()
plt.show()
```

```python
bins = 30
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=[12,6])
ax[0].hist(beta_distribution_11, bins=bins, density=density)
ax[1].hist(beta_distribution_22, bins=bins, density=density)
ax[2].hist(beta_distribution_39, bins=bins, density=density)
plt.tight_layout()
plt.show()
```

```python
# 4
bins=50
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[12,6])
ax[0].hist2d(two_dimensional_1001[:,0], two_dimensional_1001[:,1], bins=bins, density=density)
ax[1].hist2d(two_dimensional_2112[:,0], two_dimensional_2112[:,1], bins=bins, density=density)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[12,6])
sns.histplot(x=two_dimensional_1001[:,0], y=two_dimensional_1001[:,1],bins=bins, cbar=True, ax=ax[0])
sns.histplot(x=two_dimensional_2112[:,0], y=two_dimensional_2112[:,1],bins=bins, cbar=True, ax=ax[1])
plt.tight_layout()
plt.show()
```

```python
# 5
alpha = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = abs(np.random.normal(0,1,1000))

mi = alpha + beta*x
y = np.random.normal(mi, sigma)
```

```python
# 6
plt.scatter(x, y, color='r', marker='.')
for num in range(1000):
    plt.plot(np.linspace(-6,6,100), alpha[num]+beta[num]*np.linspace(-6,6,100), 'g-')
```
